```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from paddle.vision.datasets import MNIST

# Load the MNIST dataset
mnist = MNIST()
data = mnist.images
labels = mnist.labels

# Convert data to numpy array
data_np = np.array(data)

# Perform PCA dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(data_np)

# Perform Kernel PCA dimensionality reduction
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(data_np)

# Plot PCA dimensionality reduction
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral', 10))
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar()
plt.title('PCA Dimensionality Reduction')

# Plot Kernel PCA dimensionality reduction
plt.figure(figsize=(8, 6))
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=labels, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral', 10))
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar()
plt.title('Kernel PCA Dimensionality Reduction')

plt.show()
```
